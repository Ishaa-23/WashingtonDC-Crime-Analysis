```python
import pandas as pd
df=pd.read_excel('Dataset.xlsx')
df.head()
```

```python
print(df.columns)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df['END_DATE']=pd.to_datetime(df['END_DATE'])
df['REPORT_DAT']=pd.to_datetime(df['REPORT_DAT'])
df['START_DATE']=pd.to_datetime(df['START_DATE'])
df.dtypes
```

```python
df.isna().sum()
```

```python
df.drop(columns=['OCTO_RECORD_ID','BID'],inplace=True)
df.isna().sum()
```

```python
df['ANC'].fillna(df['ANC'].mode()[0],inplace=True)
df.isna().sum()
```

```python
df['WARD'].fillna(df['WARD'].mode()[0],inplace=True)
df['VOTING_PRECINCT'].fillna(df['VOTING_PRECINCT'].mode()[0],inplace=True)
df['NEIGHBORHOOD_CLUSTER'].fillna('Unknown',inplace=True)
df['BLOCK_GROUP'].fillna('Unknown',inplace=True)
df['DISTRICT'].fillna('Unknown',inplace=True)
df['CENSUS_TRACT'].fillna('Unknown',inplace=True)
df['PSA'].fillna('Unknown',inplace=True)
df['sector'].fillna('Unknown',inplace=True)
df.isna().sum()
```

```python
df.to_csv('Working dataset.csv',index=False)
```
